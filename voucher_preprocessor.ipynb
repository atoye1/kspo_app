```python
def fetch_voucher_fac_data():
    FAC_API_ENDPOINT = "http://apis.data.go.kr/B551014/SRVC_OD_API_FACIL_MNG/todz_api_facil_mng_i"
    total_fac_list = []
    for page in range(1, 100):
        print(f"Processing page {page},Processed: {len(total_fac_list)} facilities")
        fac_res = requests.get(FAC_API_ENDPOINT, params={
            "serviceKey": SERVICE_KEY,
            "resultType":"json",
            "pageNo":page,
            "numOfRows":1000,
        })
        fac_result = fac_res.json()
        fac_list = fac_result["response"]["body"]["items"]["item"]
        total_fac_list.extend(fac_list)
        if len(fac_list) < 1000:
            break
    fac_df = pd.DataFrame(total_fac_list)
    fac_df.head()
    return fac_df
```

```python
def fetch_voucer_class_data():
    CLASS_API_ENDPOINT = "http://apis.data.go.kr/B551014/SRVC_OD_API_FACIL_COURSE/todz_api_facil_course_i"
    total_class_list = []
    for page in range(1, 100):
        print(f"Processing page {page},Processed: {len(total_class_list)} classes")
        class_res = requests.get(
            CLASS_API_ENDPOINT,
            params={
                "serviceKey": SERVICE_KEY,
                "resultType": "json",
                "pageNo": page,
                "numOfRows": 1000,
            },
        )
        class_res.raise_for_status()
        class_result = class_res.json()
        class_list = class_result["response"]["body"]["items"]["item"]
        total_class_list.extend(class_list)
        if len(class_list) < 1000:
            break
    class_df = pd.DataFrame(total_class_list)
    return class_df
```

```python
fac_df = fetch_voucher_fac_data()
class_df = fetch_voucer_class_data()
len(fac_df), len(class_df)
```
